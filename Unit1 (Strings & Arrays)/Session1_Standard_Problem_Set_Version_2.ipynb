```python
def trilogy(year):
    if year == 2005:
      print("Batman Begins")
    elif year == 2008:
      print("The Dark Knight")
    elif year == 2012:
      print("The Dark Knight Rises")
    else:
      print(f"Christopher Nolan did not release a Batman movie in {year}.")

year = 2008
trilogy(year)
```

```python
def find_villain(crowd, villain):
    index = []
    for i in range(len(crowd)):
        if crowd[i] == villain:
          index.append(i)
    return index

crowd = ['Batman', 'The Joker', 'Alfred Pennyworth', 'Robin', 'The Joker', 'Catwoman', 'The Joker']
villain = 'The Joker'
find_villain(crowd, villain)
```

```python
def get_odds(nums):
    result = []
    for i in range(len(nums)):
      if nums[i] % 2 != 0:
          result.append(nums[i])

    print(result)

nums = [1, 2, 3, 4]
get_odds(nums)
```

```python
nums = [2, 4, 6, 8]
get_odds(nums)
```

```python
def up_and_down(lst):
    odd = 0
    even = 0
    for c in lst:
        if c % 2 == 0:
          even += 1
        else:
          odd += 1

    return odd - even

lst = [1, 2, 3]
up_and_down(lst)
```

```python
def running_sum(superhero_stats):
    counter = 0
    result = []
    for stat in superhero_stats:
      result.append(stat + counter)
      counter += stat

    return result

superhero_stats = [1, 2, 3, 4]
running_sum(superhero_stats)
```

```python
superhero_stats = [1, 1, 1, 1, 1]
running_sum(superhero_stats)
```

```python
superhero_stats = [3, 1, 2, 10, 1]
running_sum(superhero_stats)
```

```python
def shuffle(cards):
    result = []
    length = len(cards)
    half = len(cards) // 2
    x = cards[:half]
    y = cards[half:]
    for i in range(len(x)):
      result.append(x[i])
      result.append(y[i])

    return result


cards = ['Joker', 'Queen', 2, 3, 'Ace', 7]
shuffle(cards)
```

```python
cards = [9, 2, 3, 'Joker', 'Joker', 3, 2, 9]
shuffle(cards)
```
