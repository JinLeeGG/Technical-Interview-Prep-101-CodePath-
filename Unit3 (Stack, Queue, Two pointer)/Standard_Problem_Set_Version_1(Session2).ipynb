```python
def manage_stage_changes(changes):
    # initialize stack
    stack = []
    # initialize cancled stack
    cStack = []
    result = []
    last = []

    # loop through changes
    for change in changes:
      # if cancel, pop() the stack and append to canceld stack
      if change == "Cancel":
          cancel = stack.pop()
          cStack.append(cancel)
      # elif reschedule, pop() the canceled stack and add to stack
      elif change == "Reschedule":
          res = cStack.pop()
          stack.append(res)
      else:
        # if schedule X, append to stack
          stack.append(change)

    # this will only left with "Schedule X"
    # Seperate "Schedule" and "x"
    for s in stack:
        result.append(s.split())

    # store only x
    for i in range(len(result)):
        last.append(result[i][1])

    # return last
    return last
```

```python
print(manage_stage_changes(["Schedule A", "Schedule B", "Cancel", "Schedule C", "Reschedule", "Schedule D"]))
print(manage_stage_changes(["Schedule A", "Cancel", "Schedule B", "Cancel", "Reschedule", "Cancel"]))
print(manage_stage_changes(["Schedule X", "Schedule Y", "Cancel", "Cancel", "Schedule Z"]))
```

```python
def booth_navigation(clues):
    # initialize stack
    stack = []
    # go through clues
    for clue in clues:
      # if "back", pop() stack
      if clue == "back":
        if stack:
          stack.pop()
      else:
        # else append() clue
        stack.append(clue)


    # return stack
    return stack
```

```python
clues = [1, 2, "back", 3, 4]
print(booth_navigation(clues))

clues = [5, 3, 2, "back", "back", 7]
print(booth_navigation(clues))

clues = [1, "back", 2, "back", "back", 3]
print(booth_navigation(clues))
```

```python
def merge_schedules(schedule1, schedule2):
    result = []
    len1, len2 = len(schedule1), len(schedule2)
    min_len = min(len1, len2)

    # Merge the part where both schedules overlap
    for i in range(min_len):
        result.append(schedule1[i])
        result.append(schedule2[i])

    # Add the remaining items from the longer schedule
    if len1 > len2:
        result.extend(schedule1[min_len:])
    elif len2 > len1:
        result.extend(schedule2[min_len:])

    return "".join(result)
```

```python
print(merge_schedules("abc", "pqr"))
print(merge_schedules("ab", "pqrs"))
print(merge_schedules("abcd", "pq"))
```

```python
def next_greater_event(schedule1, schedule2):
  """
  schedule1의 각 숫자에 대해, schedule2에서 그 숫자보다 크면서 가장 작은 값을 찾습니다.

  Args:
    schedule1: 첫 번째 숫자 리스트입니다.
    schedule2: 두 번째 숫자 리스트입니다.

  Returns:
    결과를 담은 리스트입니다. 해당하는 값이 없으면 -1이 포함됩니다.
  """
  # 1. schedule2를 오름차순으로 정렬하여 검색을 용이하게 합니다.
  schedule2.sort()

  result = []

  # 2. schedule1의 각 숫자를 순회합니다.
  for num1 in schedule1:
    found_greater = False # num1보다 큰 값을 찾았는지 여부를 기록하는 플래그

    # 3. 정렬된 schedule2에서 num1보다 큰 첫 번째 숫자를 찾습니다.
    for num2 in schedule2:
      if num2 > num1:
        result.append(num2)
        found_greater = True
        # 원하는 값을 찾았으므로 더 이상 순회할 필요가 없습니다.
        break

    # 4. schedule2를 끝까지 확인했지만 큰 값을 찾지 못한 경우 -1을 추가합니다.
    if not found_greater:
      result.append(-1)

  return result
```

```python
print(next_greater_event([4, 1, 2], [1, 3, 4, 2]))
print(next_greater_event([2, 4], [1, 2, 3, 4]))
```

```python
def sort_performances_by_type(performances):
  stack = []
  # find even number in performances
  for num in performances:
    if num % 2 == 0:
      stack.append(num)

  for num in performances:
    if num % 2 != 0:
      stack.append(num)

  return stack

  # find odd number in performances
```
