# Stacks
* LIFO (Last In, First Out) data structure.
* **Use cases**: Backtracking, reversing data, expression evaluation.
* **Implementation**: Stacks are implemented using Python’s list and the append() and pop() methods.

# Queues
* FIFO (First In, First Out) data structure.
* Use cases: Task scheduling, breadth-first search (BFS).
* Implementation: Queues are implemented using Python’s deque class from the collections module.

```python
def isValid(string):
    # initialize stack
    stack = []
    if len(string) < 2:
      return False
    # loop through the string
    for char in string:
        # if its open parenthesis, push on the stack
        if char == '(':
            stack.append(char)
        # if its closed, check if the stack is empty
        if char == ')':
            # if its empty, return False
            if len(stack) == 0:
                return False
            # if its not, pop the top
            else:
                stack.pop()
    # if the stack is empty, return true
    if len(stack) == 0:
        return True
    # if not then we return false
    else:
        return False
```

```python
print(isValid("(())"))
print(isValid("(()))"))
```
