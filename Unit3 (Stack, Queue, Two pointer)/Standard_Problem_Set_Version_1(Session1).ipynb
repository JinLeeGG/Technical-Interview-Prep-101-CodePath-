```python
def is_valid_post_format(posts):
    # initialize a stack
    stack = []
    # loop through each posts:
    for post in posts:
        # if its openning "(, [, {",  add to stack
        if post == "(" or post == "[" or post == "{":
            stack.append(post)
        # if its matching closing ), ], }
        if post == ")" or post == "]" or post == "}":
            if not stack:
                return False
            a = stack.pop()
            if a == "(" and post != ")":
                return False
            if a == "[" and post != "]":
                return False
            if a == "{" and post != "}":
                return False
    if not stack:
      return True
    else:
      return False
```

```python
print(is_valid_post_format("()"))
print(is_valid_post_format("()[]{}"))
print(is_valid_post_format("(]"))
```

```python
def reverse_comments_queue(comments):
    # initialize stack
    stack = []
    # loop each comments
    for i in range(len(comments)):
      # pop each element from comments and append to stack
      stack.append(comments.pop())
    # return stack
    return stack
```

```python
print(reverse_comments_queue(["Great post!", "Love it!", "Thanks for sharing."]))
print(reverse_comments_queue(["First!", "Interesting read.", "Well written."]))
```

```python
def is_symmetrical_title(title):
    # convert title with no space and lowercase
    text = "".join(title.split()).lower()
    # initialize two pointers, first and last element in the title
    pointer1 = 0
    pointer2 = len(text)-1
    # loop through each elements in title and compare two pointers
    for _ in range(len(text)):
        # if they are not equal to each other, return false. otherwise, continue
        if text[pointer1] != text[pointer2]:
            return False
        # increment and decrement pointers
        pointer1 += 1
        pointer2 -= 1
    # if program gets through loop, return true
    return True
```

```python
print(is_symmetrical_title("A Santa at NASA"))
print(is_symmetrical_title("Social Media"))
```

```python
def engagement_boost(engagements):
    n = len(engagements)
    result = [0] * n
    left, right = 0, n - 1
    position = n - 1

    while left <= right:
        left_square = engagements[left] * engagements[left]
        right_square = engagements[right] * engagements[right]
        if left_square > right_square:
            result[position] = left_square
            left += 1
        else:
            result[position] = right_square
            right -= 1
        position -= 1

    return result
```

```python
print(engagement_boost([-4, -1, 0, 3, 10]))
print(engagement_boost([-7, -3, 2, 3, 11]))
```

```python
def clean_post(post):
  # initialize stack
  st = []
  # loop each elements in post
  for char in post:
    # if there is no element in stack, append the element
    if not st:
        st.append(char)
    # if current element in post(uppercase) and equal to last element(lowercase) in stack, pop() the stack
    if char.isupper() == True and st[-1].islower() == True and char.lower() == st[-1]:
        st.pop()
    # if current element in post(lowercase) and equal to last element(uppercase) in stack, pop() the stack
    elif char.islower() == True and st[-1].isupper() == True and char.upper() == st[-1]:
        st.pop()
    else:
      st.append(char)

  st.pop(0)
  return "".join(st)
```

```python
print(clean_post("poOost"))
print(clean_post("abBAcC"))
print(clean_post("s"))
```

```python
def edit_post(post):
  # split each word and assign to a list
  word = post.split()
  # initialize result list
  result = []

  # for each word in list
  for w in word:
      w = list(w)
      # initialize string
      string = ""
      # loop through each char in word:
      for i in range(len(w)):
          # pop() each element in word and add to string
          string += w.pop()
      # append string to stack
      result.append(string)

  return " ".join(result)
```

```python
print(edit_post("Boost your engagement with these tips"))
print(edit_post("Check out my latest vlog"))
```

```python
def post_compare(draft1, draft2):

  # make this into function
  def func(draft):
    # initialize stack
    stack = []
  # go though draft:
    for char in draft:
    # if there is no element in stack, append element
      if not stack:
          stack.append(char)
    # elif current element is backspace '#', pop() the stack
      elif char == "#":
          stack.pop()
    # otherwise, append element to stack
      else:
        stack.append(char)
  # return stack
    return stack


  if func(draft1) == func(draft2):
    return True
  else:
    return False
```

```python
print(post_compare("ab#c", "ad#c"))
print(post_compare("ab##", "c#d#"))
print(post_compare("a#c", "b"))
```
